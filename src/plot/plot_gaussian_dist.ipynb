```python
import matplotlib.pyplot as plt
import numpy as np

mean = 0
std_dev = 1
x = np.linspace(-3, 3, 1000)
y = np.exp(-(x - mean)**2 / (2 * std_dev**2)) / (np.sqrt(2 * np.pi) * std_dev)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Gaussian Distribution (μ=0, σ=1)')

plt.grid(True)
plt.xlim(-3, 3)
plt.ylim(0, 0.5)

plt.xticks(np.arange(-3, 3, step=1))
plt.yticks(np.arange(0, 0.6, step=0.1)) 

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.savefig('gaussian.png', bbox_inches='tight', dpi=150)
plt.show()
```
